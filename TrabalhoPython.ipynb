```python
#1 Importação dos pacotes:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Conjunto de dados de segmentação de imagem criado em novembro de 1990 pelo Vision Group juntamente com a Universidade de Massachusetts.

Informações sobre o conjunto de dados: As instâncias foram desenhadas aleatoriamente a partir de um banco de dados de 7 imagens externas. As imagens foram segmentadas à mão para criar uma classificação para cada pixel.

Cada instância é uma região 3x3.


```python
#3 Importação dos dados do repositório em DataFrame = df
df = pd.read_table("http://archive.ics.uci.edu/ml/machine-learning-databases/image/segmentation.data", sep=",", header=[2], skip_blank_lines=True)
display(df)
```

```python
matiz = df[["HUE-MEAN"]].dropna()
display(matiz)
```

```python
#9 Histograma dos dados do item 8
sns.displot(matiz, binwidth = 0.05, kde=True, rug=True, legend=False)
plt.title("Distribuição da Matiz")
plt.style.use('ggplot')
```
